```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTex,  tick LAVEL " )
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi / 2, np.pi],
          [r'$-\pi$', r'$-\pi/2$',r'$0$',r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],["Low","Zero","High"])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False

x1 = np.linspace(start = 0.0, stop = 5.0, num  = 50)
x2 = np.linspace(0.0, 2.0, num =50)
y1 = np.cos(2 * np.pi * x1) * np.exp(-1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title("A table of 2 subplots")
plt.ylabel("Damped oscillation")

plt.subplot(2,1,1)
plt.plot(x2,y2,'r.-')
plt.title("Undamped")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, ax = plt.subplots(2,2,figsize = (10,10))
ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)

ax[0,0].set_titile("sine Function")
ax[0,1].set_titile("sine Function")
ax[1,0].set_titile("sine Function")
ax[1,1].set_titile("sine Function")
```
