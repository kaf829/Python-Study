```python
import matplotlib.pyplot as plt
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15.55, 30, 45,9.45]
colors = ['yellowgreen', 'gold', 'lightskyblue','lightcoral']
explode = (0,0.1,0,0) #강조시키는 떨어트려서 보여줌
plt.title("pie chart")
plt.pie(sizes, explode = explode, labels =labels , colors = colors, autopct ='%1.2f%%', shadow =True, startangle = 90, counterclock = False)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.random.randn(100)
plt.title('histogram')
arrays, bins, patches = plt.hist(x, bins =10, rwidth = 1, color ='red', alpha = 0.4, density = True) #bin => 각 구간의 값, arrays => 데이터
arrays, bins, patches = plt.hist(x, bins =20, rwidth = 1, color ='blue', alpha = 0.4, density = True)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris') #titanic, tips, flights
x = iris.petal_length.values
plt.rcParams['font.family'] = 'Malgun Gothic' 
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

```python
iris = sns.load_dataset('iris') #titanic, tips, flights
x = iris.petal_length.values
plt.rcParams['font.family'] = 'Malgun Gothic' 
sns.kdeplot(x) #막대그래프를 부드럽게 표현했다고 생각하면 됨
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Desity Plot")
plt.show()
```
