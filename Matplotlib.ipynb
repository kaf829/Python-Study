```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False

x1 = np.linspace(start = 0.0, stop = 5.0, num  = 50)
x2 = np.linspace(0.0, 2.0, num =50)
y1 = np.cos(2 * np.pi * x1) * np.exp(-1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title("A table of 2 subplots")
plt.ylabel("Damped oscillation")

plt.subplot(2,1,1)
plt.plot(x2,y2,'r.-')
plt.title("Undamped")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, ax = plt.subplots(2,2,figsize = (10,10))
ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)

ax[0,0].set_title("sine Function")
ax[0,1].set_title("cos Function")
ax[1,0].set_title("tan Function")
ax[1,1].set_title("exp Function")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,y): #윤곽선이 그려지는 깊이 값 계산
    return (1-x/ 2 + x ** 5 + y **3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x,y)
ZZ = f(XX,YY)
plt.title("Control plots")
plt.contourf(XX,YY,ZZ, alpha = .75, cmap = 'hot')
plt.contour(XX,YY,ZZ, colors = "black")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = [70, 90, 80]
y2 = [60, 70, 90]

x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5
xlabel = ["국어", '영어', '수학']

plt.title("Bar Chart", fontname='Malgun Gothic')
plt.bar(x, y1, yerr=err1, ecolor="yellow", width=0.3, color='red', align='center', label='점수')
plt.bar(x + 0.4, y2, yerr=err2, width=0.3, color='blue', align='center', label='반평균')  
plt.xticks(x + 0.4 / 2, xlabel, fontname='Malgun Gothic')
plt.ylim(0, 100)
plt.xlabel("과목",fontname='Malgun Gothic')
plt.ylabel("점수",fontname='Malgun Gothic')
plt.legend()  # Corrected 'lengend' to 'legend'
plt.show()
```
