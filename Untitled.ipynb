```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import zipfile
import os

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 경로
cpi_path = "filled_state_cpi_data.csv"
zip_path = "C:/Users/MINJUN/Desktop/data/car_prices.zip"
extract_path = "C:/Users/MINJUN/Desktop/data/"

# 1. CPI 데이터 로드
cpi_df = pd.read_csv(cpi_path)
cpi_df['state'] = cpi_df['state'].str.upper()
cpi_df['year'] = cpi_df['year'].astype(int)
cpi_df['month'] = cpi_df['month'].astype(int)

# 2. car_prices.zip 압축 해제 및 데이터 로드
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
car_path = os.path.join(extract_path, "car_prices.csv")
car_df = pd.read_csv(car_path)
car_df['state'] = car_df['state'].str.upper()

# 3. 분석 대상 주
target_states = ['GA', 'NC', 'NM', 'NY']

# 4. 차량 데이터 전처리
month_map = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6,
             'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
def parse_date(row):
    try:
        parts = row['saledate'].split()
        return int(parts[3]), month_map[parts[1]]
    except:
        return None, None
car_df[['sale_year', 'sale_month']] = car_df.apply(parse_date, axis=1, result_type='expand')
car_df = car_df.dropna(subset=['sale_year', 'sale_month', 'sellingprice'])
car_df['sale_year'] = car_df['sale_year'].astype(int)
car_df['sale_month'] = car_df['sale_month'].astype(int)
car_df['sellingprice'] = car_df['sellingprice'].astype(float)
car_df = car_df[car_df['state'].isin(target_states)]

# 5. CPI 데이터에서 해당 주만 추출
cpi_df = cpi_df[cpi_df['state'].isin(target_states)]

# 6. 데이터 병합
merged = pd.merge(
    car_df,
    cpi_df,
    left_on=['state', 'sale_year', 'sale_month'],
    right_on=['state', 'year', 'month'],
    how='inner'
)

# 7. 결측치 제거 (CPI, sellingprice)
merged = merged.dropna(subset=['CPI', 'sellingprice'])

# 8. 시각화 및 분석
plt.figure(figsize=(18, 12))

# (1) CPI vs 가격 산점도
plt.subplot(2,2,1)
for state in target_states:
    temp = merged[merged['state'] == state]
    plt.scatter(temp['CPI'], temp['sellingprice'], alpha=0.4, label=state)
plt.title('CPI와 차량 가격 관계')
plt.xlabel('CPI')
plt.ylabel('가격($)')
plt.legend()

# (2) 시간별 평균 가격 추이
plt.subplot(2,2,2)
for state in target_states:
    temp = merged[merged['state'] == state]
    trend = temp.groupby('sale_year')['sellingprice'].mean()
    plt.plot(trend.index, trend.values, marker='o', label=state)
plt.title('연도별 평균 가격 추이')
plt.xlabel('연도')
plt.ylabel('평균 가격($)')
plt.legend()

# (3) CPI 변화 추이
plt.subplot(2,2,3)
for state in target_states:
    temp = merged[merged['state'] == state]
    cpi_trend = temp.groupby('sale_year')['CPI'].mean()
    plt.plot(cpi_trend.index, cpi_trend.values, marker='o', label=state)
plt.title('연도별 CPI 변화')
plt.xlabel('연도')
plt.ylabel('CPI')
plt.legend()

# (4) CPI 구간별 평균 가격 (에러 방지)
plt.subplot(2,2,4)
for state in target_states:
    temp = merged[merged['state'] == state].copy()
    temp = temp.dropna(subset=['CPI'])
    unique_vals = temp['CPI'].nunique()
    bins = min(5, unique_vals)
    if bins >= 2:
        temp['cpi_bin'] = pd.qcut(temp['CPI'], q=bins, duplicates='drop')
        price_by_bin = temp.groupby('cpi_bin', observed=True)['sellingprice'].mean()
        bin_mid = [interval.mid for interval in price_by_bin.index]
        plt.plot(bin_mid, price_by_bin.values, marker='o', label=state)
    else:
        print(f"{state}: 분위 구간(bin) 생성을 위한 데이터가 부족합니다.")
plt.title('CPI 구간별 평균 가격')
plt.xlabel('CPI 구간 중간값')
plt.ylabel('평균 가격($)')
plt.legend()

plt.tight_layout()
plt.show()

# 9. 상관관계 분석
print("주별 상관관계:")
for state in target_states:
    subset = merged[merged['state'] == state]
    corr = subset['sellingprice'].corr(subset['CPI'])
    print(f"{state}: {corr:.3f}")
```

```python
import pandas as pd
import zipfile
import os
import matplotlib.pyplot as plt


# 2. CPI 데이터 로드
cpi_df = pd.read_csv("filled_state_cpi_data.csv")
cpi_df['state'] = cpi_df['state'].str.upper()
cpi_df[['year', 'month']] = cpi_df[['year', 'month']].astype(int)

# 3. 자동차 데이터 로드 및 전처리
car_df = pd.read_csv("C:/Users/MINJUN/Desktop/data/car_prices.csv")
car_df['state'] = car_df['state'].str.upper()

# saledate 파싱 함수 수정
def parse_saledate(saledate):
    try:
        month_str, year_str = saledate.split()
        month = {
            'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6,
            'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12
        }[month_str]
        return int(year_str), month
    except:
        return None, None

# 새로운 컬럼 생성
car_df[['sale_year', 'sale_month']] = car_df['saledate'].apply(
    lambda x: pd.Series(parse_saledate(x))
)

# 결측치 제거 및 타입 변환
car_df = car_df.dropna(subset=['sale_year', 'sale_month', 'sellingprice'])
car_df['sale_year'] = car_df['sale_year'].astype(int)
car_df['sale_month'] = car_df['sale_month'].astype(int)
car_df['sellingprice'] = car_df['sellingprice'].astype(float)

# 4. 데이터 병합
merged = pd.merge(
    car_df,
    cpi_df,
    left_on=['state', 'sale_year', 'sale_month'],
    right_on=['state', 'year', 'month'],
    how='inner'
)

# 5. 분석 결과 확인
print("병합된 데이터 샘플:")
print(merged[['state', 'sale_year', 'sale_month', 'CPI', 'sellingprice']].head())

# 6. 주별 상관계수 계산
print("\n주별 CPI와 차량 가격 상관계수:")
for state in ['GA', 'NC', 'NM', 'NY']:
    temp = merged[merged['state'] == state]
    if len(temp) > 10:  # 최소 10개 데이터 점 검증
        corr = temp['CPI'].corr(temp['sellingprice'])
        print(f"{state}: {corr:.3f}")
    else:
        print(f"{state}: 분석 불가 (데이터 부족)")

# 7. 시각화
plt.figure(figsize=(12,6))
for state in ['GA', 'NC', 'NM', 'NY']:
    temp = merged[merged['state'] == state]
    if len(temp) > 10:
        plt.scatter(temp['CPI'], temp['sellingprice'], alpha=0.5, label=state)
plt.xlabel('CPI')
plt.ylabel('차량 가격($)')
plt.title('주별 CPI와 차량 가격 관계')
plt.legend()
plt.show()
```
