```python
import matplotlib.pyplot as plt
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15.55, 30, 45,9.45]
colors = ['yellowgreen', 'gold', 'lightskyblue','lightcoral']
explode = (0,0.1,0,0) #강조시키는 떨어트려서 보여줌
plt.title("pie chart")
plt.pie(sizes, explode = explode, labels =labels , colors = colors, autopct ='%1.2f%%', shadow =True, startangle = 90, counterclock = False)
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris') #titanic, tips, flights
x = iris.petal_length.values
plt.rcParams['font.family'] = 'Malgun Gothic' 
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

```python
iris = sns.load_dataset('iris') #titanic, tips, flights
x = iris.petal_length.values
plt.rcParams['font.family'] = 'Malgun Gothic' 
sns.kdeplot(x) #막대그래프를 부드럽게 표현했다고 생각하면 됨
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Desity Plot")
plt.show()
```

```python
iris = sns.load_dataset('iris') #titanic, tips, flights
x = iris.petal_length.values
plt.rcParams['font.family'] = 'Malgun Gothic' 
sns.distplot(x, vertical = True)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist plot")
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic_size = titanic.pivot_table(index = "class", columns ='sex', aggfunc = 'size')
titanic_size
```

```python
sns.heatmap(titanic_size, cmap = sns.light_palette('red', as_cmap = True), annot = True, fmt = 'd') #annot = 숫자 나오게 함 , fmt=d 숫자형
plt.title("HeatMap")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.barplot( x= 'day', y = 'total_bill', data = tips)
plt.title("요일 별, 전체 팁")
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("요일 별 전체 팁의 Box plot")
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.violinplot(y= "day", x="total_bill", data = tips, hue = 'smoker')
plt.title("요일 별 전체 팁의 Violin Plot")
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.stripplot(x = "day", y="total_bill", data = tips, jitter = True)
plt.title("요일 별 전체 팁의 Violin Plot")
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.swarmplot(x = "day", y="total_bill", data = tips)
plt.title("요일 별 전체 팁의 swarmplot ")
plt.show()
```

```python
import seaborn as sns
import numpy as np

tips = sns.load_dataset("tips")
plt.title("BoxPlot과 stripPlot으로 표현한 요일 별 팁 ")
sns.boxplot(x = "tip", y = "day", data = tips, whis = np.inf)
sns.stripplot(x = "tip", y = 'day', data = tips, jitter = True , color = '0.4')
plt.show()
```
